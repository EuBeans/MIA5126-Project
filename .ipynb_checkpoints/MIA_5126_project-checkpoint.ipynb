```python
# Use raw GitHub URL for CSV file
github_url = "https://raw.githubusercontent.com/EuBeans/MIA5126-Project/main/housing-price-dataset/crea_stats_hoursing_price.xlsx"
# Load the Excel file
xls = pd.ExcelFile(github_url)
```

```python
# Read all sheets into a dictionary of DataFrames
dfs = {}

# Find all columns that appear across sheets
all_columns = set()

for sheet in xls.sheet_names:
    df = xls.parse(sheet)
    all_columns.update(df.columns)  # Collect all unique column names

# 🔹 Process Each Sheet and Standardize Columns
for sheet in xls.sheet_names:
    df = xls.parse(sheet)

    # Ensure 'Date' column is in datetime format
    if 'Date' in df.columns:
        df['Date'] = pd.to_datetime(df['Date'])

    # Ensure all missing columns exist in this DataFrame
    for col in all_columns:
        if col not in df.columns:
            df[col] = None  # Add missing column as NaN

    # Identify HPI and Benchmark columns
    hpi_columns = [col for col in df.columns if "HPI" in col]
    benchmark_columns = [col for col in df.columns if "Benchmark" in col]

    # Convert Data Types Before Merging
    df[hpi_columns] = df[hpi_columns].astype(float)  # Convert HPI to float
    df[benchmark_columns] = df[benchmark_columns].astype(pd.Int64Dtype())  # Convert Benchmark to int (allows NaN)

    # Store cleaned DataFrame
    dfs[sheet] = df

# Check if the data is in `dfs` before merging
for city, df in dfs.items():
    if 'Townhouse_HPI' in df.columns and df['Townhouse_HPI'].isnull().all():
        print(f"🚨 Townhouse_HPI is missing in {city} BEFORE merging!")

    if 'Apartment_HPI' in df.columns and df['Apartment_HPI'].isnull().all():
        print(f"🚨 Apartment_HPI is missing in {city} BEFORE merging!")
    if 'Townhouse_Benchmark' in df.columns and df['Townhouse_Benchmark'].isnull().all():
      print(f"🚨 Townhouse_Benchmark is missing in {city} BEFORE merging!")

    if 'Apartment_Benchmark' in df.columns and df['Apartment_Benchmark'].isnull().all():
        print(f"🚨 Apartment_Benchmark is missing in {city} BEFORE merging!")


# 🔹 Merge into Long Format AFTER Fixing Missing Columns
df = pd.concat(dfs, names=["City"]).reset_index(level=0)

# 🔹 Verify Data Types Before Proceeding
print("\n✅ Final Data Types After Fixing:")
print(df.dtypes)

# 🔹 Check if missing values are still present
print("\n🚨 Missing values after merging:")
print(df.isnull().sum())

# 🔹 Save the cleaned dataset
df.to_csv("processed-datasets/corrected_housing_data.csv", index=False)

df.head()
```

#  Lag Plot Analysis of Composite HPI (Aggregate)

##  Key Observations:
- The **lag plot shows a strong linear relationship**.
- This indicates **high autocorrelation**, meaning the current HPI is highly dependent on previous values.
- The lack of random scatter suggests a **predictable trend** in housing prices.

##  What This Means:
 **High Autocorrelation** → The time series follows a clear pattern.  
 **No Random Scattering** → Past values are good predictors of future values.  
 **Forecasting is Feasible** → We can use **ARIMA** or **Prophet** for accurate predictions.

```python



# Select a specific city for analysis (e.g., Toronto)
city_name = "Aggregate"  # Change city as needed
df_city = df[df["City"] == city_name].set_index("Date")

# Visualize Composite HPI Over Time
fig = px.line(df_city, x=df_city.index, y="Composite_HPI",
              title=f"Composite HPI Trend Over Time - {city_name}",
              template="plotly_dark")
fig.show()

# Decompose the Time-Series (Trend, Seasonality, Residuals)
decomposition = sm.tsa.seasonal_decompose(df_city["Composite_HPI"], model="additive", period=12)

plt.figure(figsize=(12,8))
plt.subplot(3,1,1)
plt.plot(decomposition.trend, label="Trend")
plt.legend()

plt.subplot(3,1,2)
plt.plot(decomposition.seasonal, label="Seasonality")
plt.legend()

plt.subplot(3,1,3)
plt.plot(decomposition.resid, label="Residuals")
plt.legend()

plt.suptitle("Time-Series Decomposition")
plt.show()

# Check for Stationarity (ADF Test)
adf_test = adfuller(df_city["Composite_HPI"].dropna())
print("\n📊 ADF Test Results:")
print(f"ADF Statistic: {adf_test[0]}")
print(f"P-Value: {adf_test[1]}")
if adf_test[1] < 0.05:
    print("The data is stationary (p < 0.05).")
else:
    print(" The data is NOT stationary (p >= 0.05), differencing may be needed.")

#  Plot ACF & PACF (Autocorrelation)
fig, ax = plt.subplots(2,1, figsize=(10,6))
plot_acf(df_city["Composite_HPI"].dropna(), lags=30, ax=ax[0])
plot_pacf(df_city["Composite_HPI"].dropna(), lags=30, ax=ax[1])
plt.show()
```

# Time-Series Analysis Conclusion

## 🔹 ADF Test Results
- **ADF Statistic** = `-0.59`
- **P-Value** = `0.87`
- **Conclusion**: The data **is NOT stationary** (p ≥ 0.05).
- **Action Needed**: Apply **first-order differencing** to remove the trend.

## 🔹 ACF (Autocorrelation Function)
- The ACF shows **slow decay**, indicating a **strong trend component**.
- Confirms that the series **is non-stationary**.
- **Solution**: Differencing is required.

## 🔹 PACF (Partial Autocorrelation Function)
- The PACF **cuts off after lag 1**, suggesting an **AR(1) process**.
- **Implication**: We may use **ARIMA(p=1, d=1, q=0)** as a starting point.

## Next Steps:
1. **Apply first-order differencing (`d=1`)** to remove trends.
2. **Re-run ADF test** to confirm stationarity.
3. **Re-plot ACF/PACF** to determine ARIMA parameters (`p`, `q`).
4. **Train an ARIMA model** for forecasting.

---

```python
df_aggregate = df[df["City"] == "Aggregate"].set_index("Date")

df_aggregate["Composite_HPI_diff"] = df_aggregate["Composite_HPI"].diff()

# Drop the first row since differencing creates NaN values
df_aggregate_diff = df_aggregate.dropna()

adf_test = adfuller(df_aggregate_diff["Composite_HPI_diff"])
print("\nADF Test After First-Order Differencing:")
print(f"ADF Statistic: {adf_test[0]}")
print(f"P-Value: {adf_test[1]}")
if adf_test[1] < 0.05:
    print("The data is now stationary (p < 0.05). Ready for modeling.")
else:
    print("The data is still NOT stationary. Further differencing may be needed.")

fig, ax = plt.subplots(2,1, figsize=(12,6))

plt.subplot(2,1,1)
plot_acf(df_aggregate_diff["Composite_HPI_diff"], lags=30, ax=ax[0])
plt.title("ACF After First-Order Differencing")

plt.subplot(2,1,2)
plot_pacf(df_aggregate_diff["Composite_HPI_diff"], lags=30, ax=ax[1])
plt.title("PACF After First-Order Differencing")

plt.show()
```

```python

# Select the Aggregate Data for ARIMA
df_aggregate = df[df["City"] == "Aggregate"].set_index("Date")

# Apply First-Order Differencing for Stationarity
df_aggregate["Composite_HPI_diff"] = df_aggregate["Composite_HPI"].diff()

# Drop the first row since differencing creates NaN values
df_aggregate_diff = df_aggregate.dropna()

p, d, q = 2 ,1 ,2
print(f"\nARIMA Parameters (p, d, q): ({p}, {d}, {q})")
# Train the ARIMA Model
arima_model = ARIMA(df_aggregate["Composite_HPI"], order=(p, d, q))
arima_result = arima_model.fit()
print(arima_result.summary())
```

```python
# Plot Actual vs Forecasted HPI
residuals = pd.DataFrame(arima_result.resid)
fig, ax = plt.subplots(1,2)
residuals.plot(title="Residuals", ax=ax[0])
residuals.plot(kind='kde', title='Density', ax=ax[1])
plt.show()
```

```python
auto_arima_model = auto_arima(df_aggregate["Composite_HPI"],
                              seasonal=False,
                              stepwise=True,
                              trace=True,
                              suppress_warnings=True,
                              error_action="ignore")

# Extract best ARIMA parameters
p, d, q = 1, 1, 1
print(f"\nBest ARIMA Order: ({p}, {d}, {q})")

arima_model = ARIMA(df_aggregate["Composite_HPI"], order=(p, d, q))
arima_result = arima_model.fit()

df_aggregate["ARIMA_Fitted"] = arima_result.fittedvalues

rmse = np.sqrt(mean_squared_error(df_aggregate["Composite_HPI"].dropna(), df_aggregate["ARIMA_Fitted"].dropna()))
mape = mean_absolute_percentage_error(df_aggregate["Composite_HPI"].dropna(), df_aggregate["ARIMA_Fitted"].dropna()) * 100

print(f"\n In-Sample Forecast Evaluation:")
print(f"RMSE (Root Mean Squared Error): {rmse:.2f}")
print(f"MAPE (Mean Absolute Percentage Error): {mape:.2f}%")

plt.figure(figsize=(12,6))
plt.plot(df_aggregate["Composite_HPI"], label="Actual HPI", color='blue')
plt.plot(df_aggregate["ARIMA_Fitted"], label="ARIMA Fitted Values", color='red', linestyle="dashed")
plt.title("ARIMA In-Sample Forecasting vs Actual HPI")
plt.xlabel("Date")
plt.ylabel("HPI")
plt.legend()
plt.show()
```

## Granger Causality Results (COVID-19 Period: 2020-2022)

Granger Causality testing was conducted to analyze the causal relationship between external factors and **House Price Index (HPI)** during the COVID-19 period.

### 🔹 Key Observations

#### **Inflation Rate**
- **Lag 1:** Significant impact on HPI (`p = 0.0004`), indicating that inflation changes had an immediate effect on housing prices.
- **Lag 2 & 3:** No significant relationship (`p > 0.7`), meaning inflation's influence on HPI did not persist over multiple months.

#### **All-items (CPI)**
- **Lag 1:** Strong causality (`p = 0.0019`), showing that broad consumer price changes impacted housing prices.
- **Lag 2:** Significant (`p = 0.0161`), suggesting that the impact of CPI extended beyond the initial month.
- **Lag 3:** More pronounced effect (`p = 0.0083`), reinforcing CPI's sustained influence on HPI.

### 🔹 Interpretation

- **Inflation had an immediate effect on HPI but did not persist long-term.**
- **Broad consumer price index (All-items CPI) had both immediate and lasting effects on housing prices.**
- **The COVID-19 period likely introduced volatility, amplifying short-term economic shocks.**

```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from sklearn.linear_model import LinearRegression

# Ensure the index is a proper DateTime format
df_aggregate.index = pd.to_datetime(df_aggregate.index)

# Define forecast horizon (30 months)
forecast_horizon = 30

# Create a future date range for prediction (first day of each month)
last_date = df_aggregate.index[-1]
future_dates = pd.date_range(start=last_date + pd.DateOffset(months=1), periods=forecast_horizon, freq="MS")

# Create a placeholder for future exogenous variables
future_exog = pd.DataFrame(index=future_dates, columns=exog_vars)

future_steps = len(future_exog)  # Number of months to forecast

print("\n🔍 Injected Exogenous Spikes:")

for col in exog_vars:
    if df_aggregate[col].isnull().all():
        future_exog[col] = 0  # If the entire column is NaN, fill with zeros
    else:
        # Forecast using ARIMA
        arima_model = ARIMA(df_aggregate[col].dropna(), order=(1,1,1))
        arima_fit = arima_model.fit()
        arima_forecast = arima_fit.forecast(steps=future_steps)

        # Forecast using Linear Regression
        y = df_aggregate[col].dropna().values
        X = np.arange(len(y)).reshape(-1, 1)
        model_lr = LinearRegression()
        model_lr.fit(X, y)
        future_X = np.arange(len(y), len(y) + future_steps).reshape(-1, 1)
        lr_forecast = model_lr.predict(future_X)

        # Seasonal Moving Average
        sma_forecast = np.full(future_steps, df_aggregate[col].rolling(12, min_periods=1).mean().iloc[-1])

        # Combine forecasts with weights
        future_exog[col] = (0.5 * arima_forecast + 0.3 * lr_forecast + 0.2 * sma_forecast)

        # Introduce 3 random fluctuation spikes
        spike_indices = np.random.choice(future_exog.index, size=1, replace=False)
        spike_magnitude = 0.1 * future_exog[col].mean()  # 10% of mean value as fluctuation
        for idx in spike_indices:
            spike_value = np.random.normal(0, spike_magnitude)
            future_exog.at[idx, col] += spike_value
            print(f"  - {col} spike at {pd.to_datetime(idx).date()} (+{spike_value:.2f})")

# Fill any remaining NaN values with last known values + small random noise
for col in exog_vars:
    mean_value = df_aggregate[col].rolling(12, min_periods=1).mean().iloc[-1]
    random_noise = pd.Series(np.random.normal(0, 0.05 * mean_value, size=len(future_exog)), index=future_exog.index)
    fill_values = pd.Series(mean_value, index=future_exog.index) + random_noise
    future_exog[col] = future_exog[col].fillna(fill_values)

# Ensure future_exog matches SARIMAX expected shape
print(f"\n✅ Updated Future Exog Shape: {future_exog.shape}")  # Should be (30, len(exog_vars))

# Generate SARIMAX forecast using the updated exogenous variables
future_forecast_boxcox = sarimax_result.get_forecast(steps=forecast_horizon, exog=future_exog).predicted_mean

# Inverse Box-Cox Transformation for interpretation
future_forecast = inv_boxcox(future_forecast_boxcox, lambda_)

# Convert future_dates to DatetimeIndex to avoid plotting issues
future_forecast.index = future_dates
```

```python

df_aggregate.index = pd.to_datetime(df_aggregate.index)

# Define parameters for cross-validation
n_splits = 12  # Number of rolling windows
test_size = 12  # Test size per split (12 months)

# Store errors and forecasts
rmse_list = []
mape_list = []
forecast_series = {}

# Loop over rolling windows
for i in range(n_splits):
    train_end = len(df_aggregate) - (n_splits - i) * test_size
    test_end = train_end + test_size

    # Split train/test data
    train = df_aggregate.iloc[:train_end]
    test = df_aggregate.iloc[train_end:test_end]

    # Extract exogenous variables
    train_exog = train[exog_vars]
    test_exog = test[exog_vars]

    # Ensure test_exog matches test set size
    if len(test_exog) != test_size:
        print(f"Skipping split {i+1}: Not enough exogenous data for the test set.")
        continue

    # Train SARIMAX Model with Auto-ARIMA parameters
    sarimax_model = SARIMAX(train["Composite_HPI_boxcox"],
                            exog=train_exog,
                            order=(p, d, q),
                            seasonal_order=(P, D, Q, s),
                            trend="c",
                            enforce_stationarity=False,
                            enforce_invertibility=False)


    sarimax_result = sarimax_model.fit()

    # Predict on Test Set
    test_forecast = sarimax_result.predict(start=len(train),
                                           end=len(train) + len(test) - 1,
                                           exog=test_exog, dynamic=True)
    # **Inverse Box-Cox Transformation to Original Scale**
    test_forecast = inv_boxcox(test_forecast, lambda_)

    # Compute Errors
    rmse = np.sqrt(mean_squared_error(test["Composite_HPI"], test_forecast))
    mape = mean_absolute_percentage_error(test["Composite_HPI"], test_forecast) * 100

    # Store errors and forecasts
    rmse_list.append(rmse)
    mape_list.append(mape)
    forecast_series[f"Split {i+1}"] = test_forecast

    print(f"✅ Split {i+1}: RMSE = {rmse:.4f}, MAPE = {mape:.2f}%")

# Final evaluation
print("\n🔹 OOT Cross-Validation Results:")
print(f"📌 Average RMSE: {np.mean(rmse_list):.4f}")
print(f"📌 Average AMPE: {np.mean(mape_list):.2f}%")
```
